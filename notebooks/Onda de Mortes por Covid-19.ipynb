```python
df_world = pd.read_csv('http://datasource.coronacidades.org/world/owid/heatmap')
```

```python
br_cases = pd.read_csv('http://datasource.coronacidades.org/br/cities/cases/full')
```

```python
prepare_heatmap(df_world, place_type="country_pt", mavg_days=7, 
                title='Evolução do Número de Mortes Diárias por Covid-19 dos Países',
               datasource='Our World in Data'
               )
```

```python
prepare_heatmap(br_cases, place_type="state", mavg_days=7, 
                title='Evolução do Número de Mortes Diárias por Covid-19 por Estado',
               datasource='Brasil.io')
```

```python
br_most_pop = br_cases[
    br_cases['city_id'].isin(    
    br_cases[['city_id', 'estimated_population_2019']]
    .drop_duplicates()
    .nlargest(n=30, columns=['estimated_population_2019'])
    .city_id)]
```

```python
prepare_heatmap(br_most_pop, place_type="city", mavg_days=7, 
                title='Evolução do Número de Mortes Diárias por Covid-19 dos Municípios mais Populosos',
               datasource='Brasil.io')
```
